##unspervised model

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
ds = load_dataset("fancyzhx/ag_news")
df = pd.DataFrame(ds['train'])

texts = df['text']
print(df.head())
```

```python
import re
# Define cleaning functions
def remove_userhandles(text):
    return re.sub(r'@\w+', '', text)

def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.lower()
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply cleaning
df['clean_Text'] = df['text'].apply(remove_userhandles)
df['clean_Text'] = df['clean_Text'].apply(clean_text)


print(df[['text', 'clean_Text']].head())
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_tfidf = vectorizer.fit_transform(texts)
```

```python
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X_tfidf)
```

```python
df['Cluster'] = kmeans.labels_
print(df[['text', 'Cluster']].head())
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_tfidf.toarray())

# Step 7: Plot the clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Cluster'], cmap='viridis', s=50, alpha=0.7)
plt.title("Text Clusters Visualized using PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar()
plt.show()
```

```python
# Sample new text data (could be a new article or document)
new_text = ["Tesla launches new electric car", "Breaking news: Stock market crashes"]

# Convert new text into the same TF-IDF feature space
new_text_tfidf = vectorizer.transform(new_text)

# Predict the cluster for the new text data
predicted_clusters = kmeans.predict(new_text_tfidf)

# Print the predicted cluster labels for the new text data
print(predicted_clusters)
```

```python
# Get the feature names (words) from the TF-IDF vectorizer
terms = vectorizer.get_feature_names_out()

# Number of top terms to display for each cluster
n_top_terms = 10

# Print the top words for each cluster
for i in range(num_clusters):
    # Get the indices of the top terms for the cluster
    cluster_center = kmeans.cluster_centers_[i]
    top_indices = cluster_center.argsort()[-n_top_terms:][::-1]

    print(f"\nCluster {i} top terms:")
    for index in top_indices:
        print(f"- {terms[index]}")
```

```python
from sklearn.metrics import accuracy_score
```

```python
y = df['label']  # True labels from the dataset
y_pred = df['Predicted_Label']  # Predicted labels after cluster-to-label mapping
```

```python
score = accuracy_score(y, y_pred)
print( score)
```
